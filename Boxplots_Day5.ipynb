```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sns.set_style(style="dark", rc=None)
Phool=sns.load_dataset("iris")
Phool
sns.lineplot(x="sepal_length",y="sepal_width",data=Phool)
plt.title("Garden m Phool")
```

```python
sns.set_style(style="dark", rc=None)
Baree=sns.load_dataset("titanic")
Baree
sns.barplot(x="class",y="fare",data=Baree,
             linewidth=5, facecolor=(0.8, 0.5, 0.4, 1),
             errcolor="0.9", edgecolor=".2")
```
