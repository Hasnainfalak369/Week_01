```python
#Style with catplot
```

```python
sns.catplot(x="world_rank",y="publications",data=Ranking,color="Red")
plt.title("University's World Ranking")
sns.set_style("dark")
sns.set_style(style=None, rc=None)
plt.show
```

```python
sns.barplot(x="world_rank",y="publications",data=Ranking,linewidth=2.5,facecolor=(1,1,0,1),
           errcolor=".3",edgecolor=".2")
plt.title("University's World Ranking")
sns.set_style("dark")
sns.set_style(style=None, rc=None)
plt.show
```
