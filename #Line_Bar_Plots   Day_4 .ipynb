```python
kashti= sns.load_dataset("titanic")
kashti
plt.figure(figsize=(5,6))
sns.barplot(x="sex", y= "alone", hue=("who"), data= kashti)
plt.title("kashti")
plt.show()
```

```python
sns.barplot(x="sex", y= "alone", hue=("who"), data= kashti, order=["male","female"])
plt.title("kashti")
plt.show()
```

```python
sns.barplot(x="sex", y= "alone", hue=("who"), data= kashti, order=["male","female"], color="Green")
plt.title("kashti")
plt.show()
```

```python
sns.barplot(x="sex", y= "alone", hue=("who"), data= kashti, order=["male","female"], color="Green", ci=None)
plt.title("kashti")
plt.show()
```

```python
sns.barplot(x="sex", y= "alone", hue=("who"), data= kashti, order=["male","female"], color="Green", ci=None,palette= "colorblind")
plt.title("kashti")
plt.show()
```
